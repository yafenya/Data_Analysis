```python
cat_features = ['cut','color']
axis_features = ['meas_length','meas_width','meas_depth']
num_features = ['size','depth_percent','table_percent']
y = np.array(train.total_sales_price)
x = train.drop(columns=['total_sales_price'])
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
axis_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(missing_values=0, strategy='mean', add_indicator=False)),
    ('polynom', PolynomialFeatures(2,include_bias=False)),
    ('scaler', StandardScaler())])

clarity_map = [{
    'col':'clarity',
    'mapping':{'FL':10,
               'IF':9,
               'VVS1':8,
               'VVS2':7,
               'VS1':6,
               'VS2':5,
               'SI1':4,
               'SI2':3,
               'I1':2,
               'I2':1,
               'I3':0}
    }]

num_transformer = Pipeline(steps=[
    ('missing_num',IterativeImputer(missing_values=0,max_iter=20)),
    ('scaler',StandardScaler())])

clarity_transformer = Pipeline(steps=[
    ('ce',ce.OrdinalEncoder(mapping=clarity_map)),
    ('scaler', StandardScaler())
    ])
    
CT = ColumnTransformer([
        ("pol_std", axis_transformer, axis_features),
        ("num", num_transformer, num_features),
        ("cat", OneHotEncoder(), cat_features),
        ("ordinal_map", clarity_transformer, ['clarity'])
])
```

```python
model_target =  TransformedTargetRegressor(
                regressor = Pipeline(steps=[
                                ('preproc', CT),
                                ('estimator', Ridge())
                            ]),
                transformer=QuantileTransformer(n_quantiles=1000, output_distribution="normal")
)

model_target.fit(x_train, y_train)
```

```python
param_grid = [
    {'regressor__estimator':[KNeighborsRegressor()],
     'regressor__estimator__n_neighbors': [7,9],
      'regressor__estimator__p': [1,2],
     'regressor__preproc__num__scaler':[StandardScaler(), RobustScaler(),MinMaxScaler()],
     'regressor__preproc__pol_std__scaler' :[StandardScaler(), RobustScaler(),MinMaxScaler()],
     'transformer':[PowerTransformer(),QuantileTransformer(n_quantiles=1000, output_distribution="uniform")]},
    {'regressor__preproc__num__scaler': [StandardScaler(), RobustScaler()],
     'regressor__preproc__ordinal_map__ce':[ce.OrdinalEncoder(mapping=clarity_map), TargetEncoder()],
     'regressor__estimator':[Ridge()],
     'regressor__estimator__alpha': [0.1,1],
     'transformer':[PowerTransformer(),QuantileTransformer(n_quantiles=1000, output_distribution="uniform")]},
]
    
scores='neg_mean_absolute_percentage_error'
grid_2 = GridSearchCV(model_target, param_grid=param_grid, scoring=scores)
grid_2.fit(x_train, y_train)
```

```python
print("Лучшие параметры:", grid_2.best_params_)

# Вывести лучший результат (метрику)
print("Лучший результат (метрика):", -grid_2.best_score_)
```
